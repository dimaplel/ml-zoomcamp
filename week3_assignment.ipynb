```python
SEED = 42
```

```python
df_full_train, df_test = train_test_split(df_class, test_size=0.2, random_state=SEED)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=SEED)
assert len(df_class) == (len(df_train) + len(df_val) + len(df_test))
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = df_train['above_average']
y_val = df_val['above_average']
y_test = df_test['above_average']

# del df_train['above_average']
# del df_val['above_average']
# del df_test['above_average']
```

```python
from sklearn.metrics import mutual_info_score
```

```python
categorical = ['make', 'model', 'transmission_type', 'vehicle_style']
```

```python
def calculate_mut_inf(series: pd.Series):
    return mutual_info_score(series, df_train['above_average'])
```

```python
dv = DictVectorizer(sparse=False)

train_dict = df_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)

val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
from sklearn.metrics import accuracy_score
```

```python
np.round(accuracy_score(y_val, y_pred), 2)
```

```python
# Q4: 0.95
```

```python
orig_score = np.round(accuracy_score(y_val, y_pred), 2)
features = df_train.columns.to_list()
df_scores = pd.DataFrame(columns=['eliminated_feature', 'accuracy_score', 'accuracy_diff'])

for f in features:
    new_features = features.copy()
    new_features.remove(f)
    
    dv = DictVectorizer(sparse=False)
    
    train_dicts = df_train[new_features].to_dict(orient='records')
    X_train = dv.fit_transform(train_dicts)
    
    val_dicts = df_val[new_features].to_dict(orient='records')
    X_val = dv.transform(val_dicts)
    
    model = LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_val)
    accuracy = accuracy_score(y_val, y_pred)
    
    df_scores.loc[len(df_scores)] = [f, accuracy, orig_score - accuracy]
```

```python
df_scores
```

```python
df_scores.sort_values(by=['accuracy_diff'])
```

```python
df['price'] = np.log1p(df['price'])
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=SEED)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=SEED)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = df_train.price.values
y_val = df_val.price.values
y_test = df_test.values
```

```python
df_train = df_train.drop(['price'], axis=1)
df_val = df_val.drop(['price'], axis=1)
df_test = df_test.drop(['price'], axis=1)
```

```python
dv = DictVectorizer(sparse=False)

train_dicts = df_train.to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val.to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

```python
scores = {}
for alpha in [0, 0.01, 0.1, 1, 10]:
    model = Ridge(alpha=alpha, solver='sag', random_state=SEED)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_val)
    
    score = mean_squared_error(y_val, y_pred, squared=False)
    scores[alpha] = round(score, 3)
    print(f'alpha = {alpha}:\t RMSE = {score}')
```

```python
min(scores, key=scores.get)
```
